```python
import pandas as pd
import numpy as np

# Load the dataset. Replace with your actual file path if needed.
try:
    df = pd.read_csv(r'D:\Data Science Projects\amazon_india_project\data\amazon_india_complete_2015_2025.csv')
except FileNotFoundError:
    print("File not found. Please ensure 'amazon_india_complete_2015_2025.csv' is in the correct directory.")

# --- Perform a deep copy to avoid modifying the original DataFrame ---
df_clean = df.copy()

# ----------------------------------------
# Q1: Dates
# ----------------------------------------
# Assuming a column named 'Order Date' exists. Adjust the column name as needed.
date_col = 'Order Date'

if date_col in df_clean.columns:
    # Coerce errors to convert invalid dates to NaT (Not a Time)
    df_clean[date_col] = pd.to_datetime(df_clean[date_col], errors='coerce')
    
    # Drop rows where the date is invalid (NaT). You can also flag them instead.
    invalid_dates_count = df_clean[date_col].isnull().sum()
    if invalid_dates_count > 0:
        print(f"Warning: Dropping {invalid_dates_count} rows with invalid dates.")
        df_clean.dropna(subset=[date_col], inplace=True)
        
    # Convert to YYYY-MM-DD format. The format string '%Y-%m-%d' specifies this.
    df_clean[date_col] = df_clean[date_col].dt.strftime('%Y-%m-%d')
    
    print("Dates converted and invalid dates handled.")

# ----------------------------------------
# Q2: Prices
# ----------------------------------------
# Assuming a column named 'Price' exists. Adjust the column name.
price_col = 'Price'

if price_col in df_clean.columns:
    # First, convert the column to a string type for text manipulation.
    df_clean[price_col] = df_clean[price_col].astype(str)
    
    # Use .str.replace() with a regex pattern to remove '₹', ',' and any other non-numeric characters.
    # The regex `[^0-9.]` removes anything that is not a digit or a decimal point.
    df_clean[price_col] = df_clean[price_col].str.replace('₹', '', regex=False).str.replace(',', '', regex=False)
    
    # Then, use pd.to_numeric() to convert the cleaned strings to a numeric type.
    # 'coerce' will turn any remaining invalid values into NaN.
    df_clean[price_col] = pd.to_numeric(df_clean[price_col], errors='coerce')
    
    print("Prices cleaned and converted to numeric.")

# ----------------------------------------
# Q3: Ratings
# ----------------------------------------
# Assuming a column named 'Rating' exists. Adjust the column name.
ratings_col = 'Rating'

if ratings_col in df_clean.columns:
    # Clean the ratings column, assuming it has a format like '4.5 out of 5'.
    # This example removes the " out of 5" part and converts the result to a float.
    df_clean[ratings_col] = df_clean[ratings_col].astype(str).str.split(' ').str[0]
    df_clean[ratings_col] = pd.to_numeric(df_clean[ratings_col], errors='coerce')
    
    # Ensure all ratings are within the 1.0–5.0 range.
    df_clean[ratings_col] = df_clean[ratings_col].clip(1.0, 5.0)
    
    print("Ratings standardized to a 1.0–5.0 scale.")

# ----------------------------------------
# Q4: Cities
# ----------------------------------------
# Assuming a column named 'City' exists. Adjust the column name.
city_col = 'City'

if city_col in df_clean.columns:
    # Create a mapping dictionary for variations of city names.
    city_mapping = {
        'Bangalore': 'Bengaluru',
        'Bangaluru': 'Bengaluru',
        'Mumbai(Bombay)': 'Mumbai'
    }
    
    # Use the .replace() method to apply the mapping.
    df_clean[city_col] = df_clean[city_col].replace(city_mapping)
    
    print("City names standardized.")

# ----------------------------------------
# Q5: Booleans
# ----------------------------------------
# Assuming a column named 'In Stock' with variations like 'Yes'/'No' or 1/0.
# Adjust the column name as needed.
bool_col = 'In Stock'

if bool_col in df_clean.columns:
    # Create a mapping for truthy and falsy values.
    bool_mapping = {
        'Yes': True, 'yes': True, 1: True,
        'No': False, 'no': False, 0: False
    }
    
    # Convert to boolean using the mapping.
    df_clean[bool_col] = df_clean[bool_col].map(bool_mapping)
    
    # Handle any values that couldn't be mapped, converting them to NaN.
    df_clean[bool_col] = df_clean[bool_col].fillna(pd.NA).astype('boolean')
    
    print("Boolean column converted to True/False.")
    
# ----------------------------------------
# Q6: Categories
# ----------------------------------------
# Assuming a column named 'Category' exists.
category_col = 'Category'

if category_col in df_clean.columns:
    # Standardize by converting to a consistent case (e.g., lower) and stripping whitespace.
    df_clean[category_col] = df_clean[category_col].astype(str).str.strip().str.lower()
    
    # Example mapping for known variations, add more as needed.
    category_mapping = {
        'mobile phone': 'mobile',
        'books and media': 'books',
        'electronics appliances': 'electronics'
    }
    df_clean[category_col] = df_clean[category_col].replace(category_mapping)
    
    print("Categories standardized.")

# ----------------------------------------
# Q7: Delivery Days
# ----------------------------------------
# Assuming a column named 'Delivery Time' exists with values like '2-4 days'.
# Adjust the column name as needed.
delivery_col = 'Delivery Time'

if delivery_col in df_clean.columns:
    # Extract numeric part, handling potential missing values
    df_clean[delivery_col] = df_clean[delivery_col].astype(str).str.extract(r'(\d+)').astype(float)
    
    # Remove unrealistic values, such as negative days.
    # Set to NaN or a logical value, then handle later.
    df_clean.loc[df_clean[delivery_col] < 0, delivery_col] = np.nan
    
    print("Delivery Days converted to numeric and unrealistic values removed.")

# ----------------------------------------
# Q8: Duplicates
# ----------------------------------------
# Identify and remove duplicate rows based on all columns.
duplicates_before = df_clean.duplicated().sum()
if duplicates_before > 0:
    df_clean.drop_duplicates(inplace=True)
    print(f"Removed {duplicates_before} duplicate rows.")

# ----------------------------------------
# Q9: Outliers
# ----------------------------------------
# Fix potential decimal errors in prices (e.g., 1500.00 appearing as 150000.00).
# This is a domain-specific task; this example assumes a common pattern.
if price_col in df_clean.columns:
    # Find outliers using the IQR method.
    Q1 = df_clean[price_col].quantile(0.25)
    Q3 = df_clean[price_col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Cap outliers to prevent extreme values from skewing the data.
    df_clean[price_col] = np.where(
        df_clean[price_col] > upper_bound, upper_bound, df_clean[price_col]
    )
    df_clean[price_col] = np.where(
        df_clean[price_col] < lower_bound, lower_bound, df_clean[price_col]
    )
    
    print("Outliers in Price column handled (capped).")

# ----------------------------------------
# Q10: Payment Methods
# ----------------------------------------
# Assuming a column named 'Payment Method' exists.
payment_col = 'Payment Method'

if payment_col in df_clean.columns:
    # Standardize payment methods by converting to a consistent case and replacing variants.
    df_clean[payment_col] = df_clean[payment_col].astype(str).str.strip().str.lower()
    
    payment_mapping = {
        'cash on delivery': 'cod',
        'credit card': 'card',
        'debit card': 'card',
        'net banking': 'online',
        'upi payment': 'online'
    }
    
    df_clean[payment_col] = df_clean[payment_col].replace(payment_mapping)
    
    print("Payment Methods standardized.")

# --- Final check ---
print("\n--- Final DataFrame Info after Cleaning ---")
df_clean.info()
print("\n--- Final Cleaned DataFrame Head ---")
print(df_clean.head())

# --- Save the cleaned DataFrame to a new CSV file ---
output_filename = 'cleaned_amazon_india_2015_2025.csv'
df_clean.to_csv(output_filename, index=False)

print(f"\nCleaned data has been successfully saved to {output_filename}")
df.drop_duplicates()
df
```
